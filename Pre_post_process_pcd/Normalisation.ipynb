```python
import numpy as np
import open3d as o3d

```

```python
pcd = o3d.io.read_point_cloud(r"cup_0003_1000.ply")        # reads vertices from PLY
pts = np.asarray(pcd.points)                      # (N, 3) array
```

```python
# 1) Center at mean
mean = pts.mean(axis=0)
pts_centered = pts - mean

# 2) Radius = max distance from origin
r = np.linalg.norm(pts_centered, axis=1).max()

# 3) Scale so max radius = 1
pts_normalized_sphere = pts_centered / r

# 4) Save to PLY
pcd.points = o3d.utility.Vector3dVector(pts_normalized_sphere)
o3d.io.write_point_cloud("cup_0003_1000_normalised.ply", pcd)
```
