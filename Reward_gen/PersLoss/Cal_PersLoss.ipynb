```python
import open3d as o3d
import numpy as np
from persloss import PersistentHomologyLoss
```

```python
pcd1 = o3d.io.read_point_cloud("cube.ply")
pcd2 = o3d.io.read_point_cloud("noise_cube.ply")

downpcd1 = pcd1.voxel_down_sample(voxel_size=0.5)
downpcd2 = pcd2.voxel_down_sample(voxel_size=0.5)

pointcloud1 = np.asarray(downpcd1.points)
pointcloud2 = np.asarray(downpcd2.points)
```

```python
from persloss import PersistentHomologyLoss

# Initialize PersLoss
persloss = PersistentHomologyLoss(max_dimension=1, max_edge_length=5.0)

# Compute the loss
loss_result = persloss.compute_loss(pointcloud1, pointcloud2)

# Print results
print("PersLoss Results:")
print(f"Total Loss: {loss_result['total_loss']:.4f}")
print(f"0D Loss (connected components): {loss_result['loss_0d']:.4f}")
print(f"1D Loss (loops): {loss_result['loss_1d']:.4f}")
```
